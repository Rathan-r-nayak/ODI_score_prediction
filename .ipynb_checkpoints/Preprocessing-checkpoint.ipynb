```python
import pandas as pd
import numpy as np
from yaml import safe_load
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import pickle
```

```python
filenames = []

for file in os.listdir('data'):
    filenames.append(os.path.join('data/',file))
```

```python
final_df = pd.DataFrame()
counter = 1

#tqdm->taqadam means progress
for file in tqdm(filenames):
    with open(file,'r') as f:
        json_data = safe_load(f)
#         print("Loaded JSON data:", json_data)
        df = pd.json_normalize(json_data)
        df['match_id'] = counter
        
        final_df=pd.concat([final_df,df],ignore_index=True)
#         print(final_df)
        counter += 1
```

```python
teams=[]
batting_team=[]
bowling_team=[]
batsman=[]
bowler=[]
runs=[]
balls=[]
dissmissed=[]
city=[]
venue=[]
player_dismissed=[]
match_id=[]

for i in tqdm(range(len(df))):
    c=0
    val=df.iloc[i]['innings'][0]['1st innings']['deliveries']
    c+=1
    for deliv in range(len(val)):
        for k,v in val[deliv].items():
            balls.append(k)
            batsman.append(v['batsman'])
            bowler.append(v['bowler'])
            runs.append(v['runs']['total'])
            batting_team.append(df.iloc[i]['innings'][0]['1st innings']['team'])
            city.append(df.iloc[i][1])
            venue.append(df.iloc[i][8])
            match_id.append(df.iloc[i][9])
            teams.append(df.iloc[i][4])
            try:
                player_dismissed.append(v['wicket']['player_out'])
            except:
                player_dismissed.append('0')
```

```python
delivery_df=pd.DataFrame({
'match_id':match_id,
'teams':teams,
'batting_team':batting_team,
'ball':balls,
'batsman':batsman,
'bowler':bowler,
'runs':runs,
'player_dismissed':player_dismissed,
'city':city,
'venue':venue
})
```

```python
def bowlingTeam(row):
    for team in row['teams']:
        if(team != row['batting_team']):
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowlingTeam,axis = 1)
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum(numeric_only=False)
```

```python
df['wickets_left'] = 10-df['player_dismissed']
```

```python
df['crr']=df['current_score']*6/(df['balls_bowled'])
df.head()
```

```python
group=df.groupby('match_id')
match_id = df['match_id'].unique()
last_five = []

for id in match_id:
    valid_columns = ['runs']  # Add other valid columns if needed
    last_five.extend(group.get_group(id)[valid_columns].rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five']=last_five
```
